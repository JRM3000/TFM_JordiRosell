## 4 Dividir los datos en un conjunto de Training y uno de test 

Se procede a dividir los datos en un conjunto de Training y uno de test. Se aplica una relación **10.000 píxeles TRAINING** y el resto **TEST**

```python
import pandas as pd
import os

def splitDataFrameBlobs(fileData_Nor,fileData_OOM,typeSplit = 1, Npixel= 10000, percetTrain=60,showLogs=False):
    # typeSplit == 1 -> split for number of pixels
    # typeSplit == 2 -> split for percent

    dataBlobsOOM= pd.read_csv(fileData_OOM, sep='\t')
    dataBlobsNor= pd.read_csv(fileData_Nor, sep='\t')
    dfTrain = pd.DataFrame(columns=dataBlobsOOM.columns)
    dfTest = pd.DataFrame(columns=dataBlobsOOM.columns)
    listBlobsTrain = []
    countPixels = 0
    
    if(typeSplit==2): 
        if (percetTrain<=100) and (percetTrain>=0):
            Npixel = int((len(dataBlobsNor)/100)*percetTrain)
            typeSplit = 1
        else:
            typeSplit=-1
            
    if(typeSplit==1):

        groupsBlobs= dataBlobsOOM.groupby(["image","blob"]) #Imagen + blob
        keys = groupsBlobs.groups.keys()
        for idx, key in enumerate(keys):
            if countPixels< Npixel:
                dataBlob = groupsBlobs.get_group(key) 
                listBlobsTrain.append(key)
                dfTrain = pd.concat([dfTrain, dataBlob], ignore_index=True, sort=False)
                countPixels += len(dataBlob)
            else:
                break

        groupsBlobs= dataBlobsNor.groupby(["image","blob"]) #Imagen + blob
        keys = groupsBlobs.groups.keys()
        for idx, key in enumerate(keys):
            if not (key in listBlobsTrain):
                dataBlob = groupsBlobs.get_group(key) 
                dfTest = pd.concat([dfTest, dataBlob], ignore_index=True, sort=False)
    
    if (typeSplit==1) and showLogs:
        
        trainNBlobs= len(dfTrain.groupby(["image","blob"]).groups.keys())
        testNBlobs= len(dfTest.groupby(["image","blob"]).groups.keys())
        print("SET_TRAIN: [objects: ",trainNBlobs,", pixels: ",len(dfTrain),"]   SET_TEST: [objects: ",testNBlobs,", pixels: ",len(dfTest),"]")
                
    return dfTrain, dfTest
```
